```python
import math
import numpy
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def calcular_volumen_piramide(base, altura):
    volumen = (base ** 2 * altura) / 3
    return volumen

def calcular_volumen_cono(radio, altura):
    volumen = (math.pi * radio ** 2 * altura) / 3
    return volumen

def graficar_piramide(base, altura):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    x = [0, base, base, 0, 0]
    y = [0, 0, base, base, 0]
    z = [0, 0, 0, 0, altura]
    
    ax.plot(x, y, z)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Pir√°mide')
    plt.show()

def graficar_cono(radio, altura):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    u = numpy.linspace(0, 2 * numpy.pi, 100)
    v = numpy.linspace(0, altura, 100)

    x = numpy.outer(numpy.cos(u), numpy.ones_like(v))
    y = numpy.outer(numpy.sin(u), numpy.ones_like(v))
    z = numpy.outer(numpy.ones_like(u), v)

    x = x * radio
    y = y * radio

    ax.plot_surface(x, y, z)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Cono')
    plt.show()

# Ejemplo de uso para una pir√°mide con base de 5 y altura de 8
base = 1
altura = 1
volumen_piramide = calcular_volumen_piramide(base, altura)
print("El volumen de la pir√°mide es:", volumen_piramide)
graficar_piramide(base, altura)

# Ejemplo de uso para un cono con radio de 3 y altura de 6
radio = 3
altura = 6
volumen_cono = calcular_volumen_cono(radio, altura)
print("El volumen del cono es:", volumen_cono)
graficar_cono(radio, altura)
```
