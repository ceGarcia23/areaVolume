```python
import matplotlib.pyplot as plt

def area_trapecio(base_mayor, base_menor, altura):
    return (base_mayor + base_menor) * altura / 2

def graficar_trapecio(base_mayor, base_menor, altura):
    # Definir las coordenadas de los vértices del trapecio
    x = [0, base_menor, base_mayor, 2*base_mayor-base_menor, 0]
    y = [0, altura, altura, 0, 0]
    
    # Graficar el trapecio
    plt.plot(x, y)
    plt.title("Trapecio Isósceles")
    plt.xlabel("Base")
    plt.ylabel("Altura")
    plt.show()

print("Cálculo de área y gráfica de un trapecio isósceles")
base_mayor = float(input("Ingrese la longitud de la base mayor: "))
base_menor = float(input("Ingrese la longitud de la base menor: "))
altura = float(input("Ingrese la altura del trapecio: "))

area = area_trapecio(base_mayor, base_menor, altura)
print("El área del trapecio es:", area)

graficar_trapecio(base_mayor, base_menor, altura)
```
