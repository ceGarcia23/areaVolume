```python
import matplotlib.pyplot as plt

def area_trapecio(base_mayor, base_menor, altura):
    area = ((base_mayor + base_menor) / 2) * altura
    return area

def graficar_area_trapecio(base_mayor, base_menor, altura):
    area = ((base_mayor + base_menor) / 2) * altura
    x = [0, base_menor, base_mayor, 0]
    y = [0, altura, altura, 0]
    plt.fill(x, y, 'g', alpha=0.3)
    plt.plot([0, base_menor], [0, 0], 'k')
    plt.plot([base_menor, base_mayor], [altura, altura], 'k')
    plt.plot([base_mayor, 0], [0, 0], 'k')
    plt.text(base_menor/2, -0.5, f"{base_menor}", ha="center")
    plt.text(base_mayor/2, altura+0.5, f"{base_mayor}", ha="center")
    plt.text(-0.5, altura/2, f"{altura}", va="center", rotation=90)
    plt.axis('equal')
    plt.show()
    return area

def main():
    B = float(input("Ingrese la longitud de la base mayor del trapecio: "))
    b = float(input("Ingrese la longitud de la base menor del trapecio: "))
    h = float(input("Ingrese la altura del trapecio: "))

    area = area_trapecio(B, b, h)
    print("El Ã¡rea del trapecio es:", area)

    graficar_area_trapecio(B, b, h)

if __name__ == "__main__":
    main()
```
