```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np
def calcular_volumen(a, b, c):
    volumen = a * b * c
    return volumen
def graficar_ortoedro(a, b, c):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    vertices = np.array([[0,0,0],[0,b,0],[a,0,0],[a,b,0],
                         [0,0,c],[0,b,c],[a,0,c],[a,b,c]])
    caras = np.array([[0,1,3,2],[0,4,5,1],[0,2,6,4],[1,5,7,3],[2,3,7,6],[4,6,7,5]])
    ortoedro = Poly3DCollection(vertices[caras], alpha=0.25, facecolor='blue', edgecolor='black')
    ax.add_collection3d(ortoedro)
    ax.set_xlim([0, a])
    ax.set_ylim([0, b])
    ax.set_zlim([0, c])
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()
a = float(input("Ingrese la medida del lado a del ortoedro: "))
b = float(input("Ingrese la medida del lado b del ortoedro: "))
c = float(input("Ingrese la medida del lado c del ortoedro: "))
volumen = calcular_volumen(a, b, c)
print(f"El volumen del ortoedro es: {volumen}")
graficar_ortoedro(a, b, c)
```
