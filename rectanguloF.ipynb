```python

import matplotlib.pyplot as plt

# Pedir al usuario los datos del rectángulo
longitud = float(input("Introduce la longitud del rectángulo: "))
altura = float(input("Introduce la altura del rectángulo: "))

# Calcular el área del rectángulo
area = longitud * altura

# Crear el gráfico del rectángulo
fig, ax = plt.subplots()
ax.set_aspect("equal")
rect = plt.Rectangle((0, 0), longitud, altura, linewidth=1, edgecolor="r", facecolor="none")
ax.add_patch(rect)

# Añadir etiquetas al gráfico
ax.set_title(f"Área del rectángulo = {area}")
ax.set_xlabel("Longitud")
ax.set_ylabel("Altura")

# Mostrar el gráfico
plt.show()
```
