```python
import matplotlib.pyplot as plt

def graficar_trapecio(base_mayor, base_menor, altura):
    # Definir las coordenadas de los vértices del trapecio
    x = [0, base_menor, base_mayor, 0]
    y = [0, altura, altura, 0]
    
    # Graficar el trapecio
    plt.plot(x, y)
    plt.title("Trapecio")
    plt.xlabel("Base")
    plt.ylabel("Altura")
    plt.show()
    
graficar_trapecio(8, 4, 6)
```
