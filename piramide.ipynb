```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
def calcular_volumen_piramide(lado_base, altura):
    area_base = lado_base ** 2
    volumen = (1/3) * area_base * altura
    return volumen
def graficar_piramide(lado_base, altura):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    # Coordenadas de los v√©rtices de la pir√°mide
    vertices = np.array([[0, 0, 0], [lado_base, 0, 0], [lado_base, lado_base, 0], [0, lado_base, 0], [lado_base/2, lado_base/2, altura]])
    # Caras de la pir√°mide
    caras = [[vertices[0], vertices[1], vertices[4]],
             [vertices[1], vertices[2], vertices[4]],
             [vertices[2], vertices[3], vertices[4]],
             [vertices[3], vertices[0], vertices[4]],
             [vertices[1], vertices[0], vertices[3], vertices[2]]]
    # Crear una colecci√≥n de pol√≠gonos 3D y agregarlos al gr√°fico
    poligonos = Poly3DCollection(caras, linewidths=1, edgecolors='k')
    poligonos.set_facecolor((0, 0.5, 0.5, 0.5))  # color de las caras
    ax.add_collection(poligonos)
    # Etiquetas de los ejes
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()
# Ejemplo de uso
lado = 1
altura = 1
volumen_piramide = calcular_volumen_piramide(lado, altura)
print("El volumen de la pir√°mide es:", volumen_piramide)
graficar_piramide(lado, altura)
```
